```python
from sklearn.datasets import fetch_openml

```

```python
dataset = fetch_openml("mnist_784")
dataset
```

```python
X = dataset.data
y = dataset.target
```

```python
numero_aleatorio = X.sample(random_state=52)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

floresta = RandomForestClassifier().fit(X_train, y_train)   
```

```python
y_pred = floresta.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)*100
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

matrizinha = confusion_matrix(y_test, y_pred)
sns.heatmap(matrizinha, annot=True, fmt="d")
```
